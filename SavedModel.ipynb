# Imports

```python
import numpy as np
import tensorflow.compat.v1 as tf
from tensorflow.compat.v1.saved_model import load, tag_constants
```

# Constants

```python
IMAGE_WIDTH = 28
IMAGE_HEIGHT = 28
CHANNELS = 1
TOTAL_INPUTS = IMAGE_WIDTH*IMAGE_HEIGHT*CHANNELS
```

# Load the Data

```python
from keras.datasets import mnist
(x_train_all, y_train_all),(x_test, y_test) = mnist.load_data()
```

```python
x_train_all = x_train_all.reshape((x_train_all.shape[0], TOTAL_INPUTS)).astype('float32')
x_test = x_test.reshape((x_test.shape[0], TOTAL_INPUTS)).astype('float32')
x_train_all.shape
x_test.shape
```

# Create Session

```python
graph = tf.Graph()
sess = tf.Session(graph = graph)
```

# Load the Model

```python
tf.saved_model.load(sess,
                    [tag_constants.SERVING],
                    export_dir="SavedModel")
```

# Run Session & Make Prediction

```python
prediction = sess.run(fetches='accuracy/Model_prediction:0' ,feed_dict={'X:0':x_test})
```

# Check Results

```python
y_test[:5]
```

```python
prediction[:5]
```

```python
nr_correct =sum(np.equal(prediction,y_test))
nr_correct/len(prediction)
```

# Closing Session


```python
sess.close()
tf.reset_default_graph
```
